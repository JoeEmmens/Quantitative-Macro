```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
UKdata = pd.read_csv("UK Emp Rate.csv").set_index("Title")
```

```python
dates = pd.Series(pd.date_range(start='2018/01/01', periods=len(UKdata.index), freq='M'))
```

```python
fig, ax = plt.subplots(facecolor="w")
ax.plot(dates, UKdata["Employment rate"], label="Total")
ax.plot(dates, UKdata["Female employment rate"], label="Female")
ax.plot(dates, UKdata["Male employment rate"], label="Male")
plt.legend()
plt.title("UK Employment Rate")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("UK Employment Rate")
```

```python
fig, ax = plt.subplots(facecolor="w")
ax.plot(dates, UKdata["Avg  weekly hours total"], label="Total")
ax.plot(dates, UKdata["Avg  weekly hours female"], label="Female")
ax.plot(dates, UKdata["Avg  weekly hours male"], label="Male")
plt.legend()
plt.title("UK Average Hours")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("UK Average Hours")

```
